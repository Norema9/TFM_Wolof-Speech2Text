```python
def extract_first_page(filename):
    pdf = PdfReader(filename)
    pdf_writer = PdfWriter()
    pdf_writer.add_page(pdf.pages[8])

    with open('first_page.pdf', 'wb') as out:
        pdf_writer.write(out)
extract_first_page('bataaxal bu gudde nii.pdf')
```

```python
def convert_pdf_to_image(pdf_path, output_path, page):
    with Image(filename=pdf_path) as img:
        img.save(filename=f'{output_path}_{page}.png')

convert_pdf_to_image('first_page.pdf', 'output_image.png', 3)
```
